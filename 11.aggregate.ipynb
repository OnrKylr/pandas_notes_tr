```python
import pandas as pd
import numpy as np
df = pd.DataFrame({'gruplar': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'degisken1': [10,23,33,22,11,99],
                   'degisken2': [100,253,333,262,111,969]},
                   columns = ['gruplar', 'degisken1', 'degisken2'])
df
```

```python
df.groupby("gruplar").mean()
```

```python
df.groupby("gruplar").aggregate(["min" , np.median ,max]) #pandas kendi fonksiyonlarini tirnakli ya da tirnaksiz alabilir anlar ama dis fonksiyonlar (buraadaki np.median) tirnaksiz olmali
```

```python
df.groupby("gruplar").aggregate({"degisken1": "min" , "degisken2": "max"})
```
